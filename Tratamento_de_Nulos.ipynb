```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame([(0.0, np.nan, -1.0, 1.0),
                   (np.nan, 2.0, np.nan, np.nan),
                   (2.0, 3.0, np.nan, 9.0),
                   (np.nan, 4.0, -4.0, 16.0)],
                  columns=list('abcd'))
df
```

# Fill Value

* Cria-se uma vaiável, nesse caso a de nome "fill_value" e se atribui o valor o qual estará substituindo os nulos na colula escolhida por meio da função "fillna()"

```python
fill_value = df.b.mean() #nesse caso o valor atribuido foi a média da coluna "b"
df['b'].fillna(fill_value)
```

```python
#jogando o resultado no df
fill_value = df.b.mean()
df['b'] = df['b'].fillna(fill_value)
df
```
